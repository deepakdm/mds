```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'LondON_StockhOlm', 'Budapest_PaRis', 'Brussels_londOn'], 
                   'FlightNumber': [10045, np.nan, 10065, np.nan, 10085], 
                   'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]], 
                   'Airline': ['KLM(!)', '<Air France>(12)', '(British Airways. )', '12. Air France', '"Swiss Air"']})
```

```python
# question 1
df['FlightNumber'] = df['FlightNumber'].interpolate().astype(int)
df
```

```python
# question 2
df['From'] = df['From_To'].str.split('_').str[0]
df['To'] = df['From_To'].str.split('_').str[1]
df
```

```python
# question 3
df['From'] = df['From'].str.capitalize()
df['To'] = df['To'].str.capitalize()
df
```

```python
# question 4
df.drop(labels='From_To', axis=1, inplace=True)
df
```

```python
# question 5
time_delay = pd.DataFrame(df.RecentDelays)
time_delay = pd.DataFrame(df['RecentDelays'].values.tolist())
time_delay.columns = ['Delay_1', 'Delay_2', 'Delay_3']
df = df.drop(labels='RecentDelays', axis=1)
df.insert(4, "Delay_1", time_delay['Delay_1'])
df.insert(5, "Delay_2", time_delay['Delay_2'])
df.insert(6, "Delay_3", time_delay['Delay_3'])
df
```
