```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def pie_chart(labels, sizes, explode):
    fig, ax = plt.subplots()
    ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
    ax.axis('equal')
    plt.show()
```

```python
def scatter_chart(x_male, y_male, x_female, y_female):
    s = 100
    plt.scatter(x_male, y_male, s, c="b", alpha=0.5, label="Male")
    plt.scatter(x_female, y_female, s, c="r", alpha=0.5, label="Female")
    plt.xlabel("Age")
    plt.ylabel("Fare")
    plt.legend(loc=2)
    plt.show()
```

```python
# Downloaded dataset from https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv
dataset = pd.read_csv('titanic_original.csv')
df = dataset.dropna(subset=['sex'])
```

```python
m_cnt = df.groupby(['sex']).get_group('male').shape[0]
f_cnt = df.groupby(['sex']).get_group('female').shape[0]
pie_chart(['Male', 'Female'],[m_cnt, f_cnt], (0.0,0.0))
```

```python
m_ds_age = df.groupby(['sex']).get_group('male').loc[:, 'age']
m_ds_fare = df.groupby(['sex']).get_group('male').loc[:, 'fare']
f_ds_age = df.groupby(['sex']).get_group('female').loc[:, 'age']
f_ds_fare = df.groupby(['sex']).get_group('female').loc[:, 'fare']
scatter_chart(m_ds_age, m_ds_fare, f_ds_age, f_ds_fare)
```
