```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Boston House Dataset
from sklearn.datasets import load_boston
boston = load_boston()
bos = pd.DataFrame(boston.data)
tar = pd.DataFrame(boston.target)
```

```python
# Split data into Train set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(bos, tar, test_size=1/4, random_state=0)
```

```python
# Apply Simple Linear Regression
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr.fit(X_train, y_train)
predicted_data = slr.predict(X_test)
expected_data = y_test
```

```python
# Plot a diagram
plt.figure(figsize=(16, 9))
plt.scatter(expected_data, predicted_data, color='r')
plt.plot([0, 50], [0, 50], '--')
plt.axis('tight')
plt.xlabel('Actual price')
plt.ylabel('Predicted price')
plt.tight_layout()
```

```python
import numpy as np
import scipy.stats as stats
print("RMS: %r " % np.sqrt(np.mean((predicted_data - expected_data) ** 2)))
```

```python
# Plot all features against Price
for index, feature_name in enumerate(boston.feature_names):
    plt.figure(figsize=(5,4))
    plt.scatter(boston.data[:, index], boston.target)
    plt.ylabel('Price', size=15)
    plt.xlabel(feature_name, size=15)
    plt.tight_layout()
```
